```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd
import logging


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

# train data
train_df = pd.read_csv('data/trainingST3.csv')
train_df = train_df[['text','labels']].values.tolist() # trasformo i records in liste per utilizzare SimpleTransformers
    

# eval data

eval_df = pd.read_csv('data/testST3.csv')
eval_df = eval_df[['text','labels']].values.tolist() # trasformo i records in liste per utilizzare SimpleTransformers

#  model configuration
model_args = ClassificationArgs(num_train_epochs=20, overwrite_output_dir = True)

# ClassificationModel
model = ClassificationModel(
    "bert", "m-polignano-uniba/bert_uncased_L-12_H-768_A-12_italian_alb3rt0", args=model_args, use_cuda =False
)

# Train the model
model.train_model(pd.DataFrame(train_df))

# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(pd.DataFrame(eval_df))

```
